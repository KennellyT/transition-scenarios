```python
import numpy as np
import matplotlib.pyplot as plt

import cymetric
import transition_metrics as tm
```

## Comparing LEU Enrichment Facility Caps

```python
# No cap on Fabrication Facility Cap
caps = [70000000, 60000000, 50000000, 10000000, 5000000, 2000000,
        1500000, 1000000, 500000, 100000, 10000, 1000, 100, 10, 1, 0]

plt.figure()
for cap in caps:
    file = 'united_states_xe100_' + str(cap) + '_None.sqlite'
    energy = tm.get_electricity(file)
    plt.plot(energy['Year'], energy['Energy'], label= str(cap))
    
plt.xlabel('Year')
plt.ylabel('Energy (MWe-y)')
plt.legend(loc=(1.04,0))
plt.title('Effect of Enrichment Cap, no fabrication cap')
```
